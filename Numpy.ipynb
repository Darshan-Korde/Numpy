```python
import numpy as np
```

```python
arr = np.array([20,10,5,4,78,89])
arr2 = np.array([[[[1,2,3,4,5],[4,5,6,7,8],[5,6,7,8,9]]]])
arr3 = np.array([[[1,2,3,4],[4,5,6,7]],[[1,2,3,6],[6,4,5,8]]])
print(arr)
print(type(arr))
print(arr2.ndim)
print(arr3[0,1,1])
print(arr2[0,0])
```

```python
print(arr[1:3])
print(arr3[0:2:2])
```

```python
print(arr2.dtype)
#data types --> i,b,f,u,c,M(datetime),m(timedata),O,S,U(unicode string)
# converting data type on existing array use astype() func
print(arr.astype('f'))
```

```python
x = arr.copy()
arr[0]=52    #copy not affect any changes made to original array
print(arr)
print(x)

y = arr.view()
arr[1]=65     # view affect if any cahnges made to original array
print(arr)
print(y)
# copy own the data and view doesn't to check use base attribute
print(x.base) #if owns data --> None
print(y.base) #if owns data --> ret original array
```

```python
newarr = arr.reshape(2,3)
print(newarr)     # 1D to 2D

newarr2 = arr3.reshape(2,2,4)  # 1D to #3D
print(newarr2)

newarr3 = arr.reshape(2,-1)  # if unknown the dimension use -1 as 2nd argument it automatically set
print(newarr3)

newarr4 = arr2.reshape(-1)
print(newarr4)  # flatten array   adv method --> flatten, rot90, flip,fliplr, flipud
```

RANDOM

```python
import numpy as np
from numpy import random
x = random.randint(10) # random no generate
x2 = random.rand(4) # list of four float no
x3 = random.randint(25, size=(3,2))  # list of 5 no
x4 = random.choice([2,5,6,8,9,1], p=[0.1,0.1,0.2,0.2,0.3,0.1], size=(2,2)) # random data distribution based on probability of element occur inn array
arr = np.array([2,5,4,8,9])
random.shuffle(arr) # change the original array
random.permutation(arr)  # unchange original array
print(arr)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.title('Normal Distributon')
sns.displot(random.normal(loc=2,scale=4, size=(100)),kind='kde' )
plt.title('Binomial Distribution')
sns.displot(random.binomial(n=5,p=0.6, size=100),kind='kde')
plt.title('Poisson Distribution')
sns.displot(random.poisson(lam=2, size=100),kind='kde')
# uniform distribution --> low (0.0)  high (1.0), scale
# logistic distribution  --> loc, sclae, size loc --> def 0 scale --> def 1  usee in neural ntwk, logistic regression
# exponential distribution  --> loc, size  loc --> def 1
# chi-square distribution --> df, size
plt.show()
```

UFUNCS --> Universal Functions


```python
import numpy as np
x = [1,1,12,15]
y= [7,8,9,11]
z=[]
for i,j in zip(x,y): # without ufuncs
  z.append(i+j)
print(z)

w = np.add(x,y)  # using ufunc
print(w)

#Arithmetic ufunc --> add, subtract, multiply, divide, power, absolute, mod, remainder, divmod --> quotient and mod
#Rounding --> trunc, fix (close to zero remove decimal), around, floor, ceil
#log func --> log2(), log10(), log()  nplog = np.frompyfunc(log, 2, 1) --> at any base
#summations --> sum, cumsum
#products --> prod, cumprod
#differences --> diff
#lcm --> lcm.reduce(arr)   reduce use when lcm of array need to find bcoz array ia 1D we want o/p 0D
#gcd --> gcd.reduce()
#trignometric --> sin,cos,tan  np.deg2rad  <--> np.rad2deg, hypot(base,perp), sinh,cosh,tanh
# set --> unique,union1d, intersect1d, setdiff1d
```
